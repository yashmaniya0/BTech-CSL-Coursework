## Q1

```python
r = 3

def alpha_func_1(r,k=0):
    return 1/5
def alpha_func_2(r,k):
    return 1 / (np.abs(r - 5) + k)

def subg_descent(x, f, alpha_func,subg_func, max_itr=500):
    func = f(x)
    alpha_values,func_values = [] , []
    for k in range(1, max_itr + 1):
        alpha = alpha_func(r,k)
        alpha_values.append(alpha)
        subg = subg_func(x)
        x = x - alpha * subg
        fcurrent = f(x)
        if fcurrent < func:
            func = fcurrent
        func_values.append(func)
    return func_values

def f(x):
    return max((x[0] - 2)**2 + (x[1] + 2)**2, x[0]**2 + 8*x[1])

def subg(x):
    if (x[0] - 2)**2 + (x[1] + 2)**2 > x[0]**2 + 8*x[1]:
        return np.array([2*(x[0] - 2), 2*(x[1] + 2)])
    elif (x[0] - 2)**2 + (x[1] + 2)**2 < x[0]**2 + 8*x[1]:
        return np.array([2*x[0], 8])
    else:
        t = np.random.uniform(0,1)
        return t*np.array([2*x[0], 8]) + (1-t)*np.array([2*(x[0] - 2), 2*(x[1] + 2)])
x = np.array([0, 0])
result_alpha_1 = subg_descent(x, f, alpha_func_1, subg)
result_alpha_2 = subg_descent(x, f, alpha_func_2, subg)


print("Result 1: ", result_alpha_1)
print("Result 2: ", result_alpha_2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(result_alpha_1) + 1), result_alpha_1, label='Subgradient Descent')
plt.xlabel('Iteration')
plt.ylabel('fbest')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(result_alpha_2) + 1), result_alpha_2, label='Subgradient Descent')
plt.xlabel('Iteration')
plt.ylabel('fbest')
plt.legend()
plt.show()
```

## Q2

```python

def f(x):
    return max((x[0] )**2 + (x[1] )**2, 5*x[0] + x[1])

def helper(x):
    if (x[0] )**2 + (x[1] )**2 > 5*x[0] + x[1]:
        return np.array([2*(x[0]), 2*(x[1])])
    elif (x[0] )**2 + (x[1] )**2 < 5*x[0] + x[1]:
        return np.array([5, 0])
    else:
        t = np.random.uniform(0,1)
        sub1 = np.array([2*(x[0]), 2*(x[1])])
        sub2 = np.array([5, 0])
        return t*sub1 + (1-t)*sub2

x = np.array([2.5, 2.5])
res_1 = subg_descent(x, f, alpha_func_1, helper)
res_2 = subg_descent(x, f, alpha_func_2, helper)


print("Result 1: ", res_1)
print("Result 2: ", res_2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(res_1) + 1), res_1, label='Subgradient Descent')
plt.xlabel('Iteration')
plt.ylabel('fbest')
plt.legend()
plt.show()
```
