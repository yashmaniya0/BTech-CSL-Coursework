#Story 11

```python
df=pd.read_csv("/content/13_AdjacencyDirectedWeighted.csv",sep=" ")
df.head()
```

```python
sns.heatmap(df)
```

```python
a=df.columns.to_list()
ab=10*a
```

```python
df_melted = pd.melt(df, value_vars=df.columns)
df_melted.rename(columns = {'variable':'source'}, inplace = True)
df_melted['target']=ab
df_melted.head()
```

```python
for i in a:
    df_melted=df_melted.replace(i,a.index(i))
df_melted.head()
```

```python
import plotly.graph_objects as go

# Define the data for the chart
data = dict(
    type='sankey',
    node = dict(pad = 15,thickness = 20,line = dict(color = "black",width = 0.5),
    label = ['Africa', 'East Asia','Europe','Latin America','North America','Oceania','South Asia','South East Asia','Soviet Union','West Asia'],
    color = ["blue", "red", "green", "black", "cyan", "pink","orange","yellow","maroon","lime"]
    ),
    link = dict(
      source = df_melted['source'],
      target = df_melted['target'],
      value = df_melted['value']
  ))
layout =  dict(
    title = "Basic Sankey Diagram",
    font = dict(
      size = 10
    )
)
fig = go.Figure(data=[data], layout=layout)

# Show the figure
fig.show()
```

```python
df=pd.read_csv("/content/10_OneNumSevCatSubgroupsSevObs.csv", sep=",")
df["tipPercent"]=[100*df['tip'][i]/df['total_bill'][i] for i in range(len(df))]
df.head()
```

```python
df["day"] = df["day"].astype("category")
df["day"].cat.reorder_categories(["Thur", "Fri", "Sat", "Sun"], inplace=True)
df["sex"] = df["sex"].astype("category")
df["sex"].cat.reorder_categories(["Male", "Female"], inplace=True)
df["time"] = df["time"].astype("category")
df["time"].cat.reorder_categories(["Lunch", "Dinner"], inplace=True)
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x="day", y="tipPercent", hue="sex", data=df, dodge=True, width=0.5, fliersize=0)
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(x="sex", y="tipPercent",hue="day",data=df,dodge=True, width=0.5, fliersize=0)
```

```python
plt.figure(figsize=(12,10))
sns.violinplot(x="day", y="tipPercent",hue="sex",data=df,dodge=True, width=0.5)
```

```python
plt.figure(figsize=(12,10))
sns.violinplot(x="sex", y="tipPercent", hue="day",data=df,dodge=True, width=0.5)
```

```python
a = sns.FacetGrid(data=df, row="sex", col="time")
a.map(sns.histplot, "tipPercent", bins=20, kde=False, fill="#69b3a2", color="black")
```
