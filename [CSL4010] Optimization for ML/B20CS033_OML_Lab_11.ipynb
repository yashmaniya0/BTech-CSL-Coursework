```python
from gekko import GEKKO
from numpy import *
```

# 1.1

```python
m = GEKKO(remote=False)
c = array([[9],[5],[6],[4]])
A = array([[6,3,5,2],[1,0,0,1],[-1,0,1,0],[0,-1,0,1]])
b = array([[10],[1],[0],[0]])
```

```python
z = m.Array(m.Var,4,integer=True,lb=0,ub=1)
m.qobj(c,x=z,otype='max')
m.axb(A,b,x=z,etype='<=')
m.options.SOLVER = 1
m.solve()
```

```python
print('Objective: ', m.options.OBJFCNVAL)
print(z)
```
