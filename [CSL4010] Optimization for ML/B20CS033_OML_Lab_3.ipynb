## Q1

Price of house (in Rs) with respect to area (sq feet) is give in 2 column excel sheet. Using linear regression, find the price of a
house with R-thousand square feet where R is the last digit of your roll no. If last digit of your roll no is 0, choose R = 2.5.


```python
df = pd.read_csv("C:/Users/YASH MANIYA/Desktop/Python/OML Labs/2_col.csv")
x = np.array(df.iloc[:, 0])
y = np.array(df.iloc[:, 1])
```

```python
A = np.column_stack((x, np.ones((len(x), 1), dtype=float)))
beta = np.dot(np.linalg.inv(np.dot(A.T, A)), np.dot(A.T, y.T))
print(beta)
```

```python
plt.figure(figsize=(10, 8))
plt.title('Price vs Area', fontsize = 20)
plt.plot(x, y, '.', markersize = 10)
plt.plot(x, beta[0]*(x)+beta[1], color='orange', linewidth=3)
plt.xlabel('Area', fontsize = 15)
plt.ylabel('Price', fontsize = 15)
plt.show()
```

```python

print('\nThe price of a house with 3000 square feet \n')
print(f'Predicted Price : Rs {beta[0]*(R*1000)+beta[1]:.4f}')
```

## Q8

Home work Define a function polynomial f it(x, y, p) to generate p degree polynomial. Using this justify which degree
polynomial gives best approximation among 1 to 20 degree polynomials i.e. which degree polynomial has lowest average loss.


```python
def polynomial_fit(x, y, p):
    l = []
    for i in range(p, -1, -1):
        l.append((x**i))
    A = np.column_stack(tuple(l))
    beta = np.dot(np.linalg.inv(np.dot(A.T, A)), np.dot(A.T, y.T))
    return A, beta
```

```python
data = pd.read_csv("C:/Users/YASH MANIYA/Desktop/Python/OML Labs/2_col.csv")
x = np.array(data.iloc[:, 0])
y = np.array(data.iloc[:, 1])
```

```python
area = x
price = y
print(area.min(), area.max())
```

```python
arr_loss = []
arr_rmse = []
for p in range(1, 21):
    A, beta = polynomial_fit(area, price, p)
    mat = np.dot(A, beta)
    mat = mat - price
    avg_loss = np.linalg.norm(mat)
    avg_loss = avg_loss/(2+len(price))
    mse = 0
    for j in range(len(area)):
        ans = 0
        for i in range(len(beta)):
            ans += beta[i]*(x[j]**(p-i))
        pred_price = ans
        mse += (pred_price-price[j])**2

    mse = mse/len(area)
    rmse = np.sqrt(mse)

    arr_rmse.append(rmse)
    arr_loss.append(avg_loss)
    print("The average loss for degree -", p, " is : ", avg_loss)
    print("The root mean squared error for degree -", p, " is :", rmse)
    plt.scatter(area, price)
    x1 = np.linspace(area.min(), area.max()+0.01)
    plt.plot(x1, np.polyval(beta, x1), 'r')
    plt.xlabel('area')
    plt.ylabel('price')
    plt.title("degree = {}". format(p))
    plt.show()
```

```python
print(f"{1+np.argmin(arr_loss)} Degree polynomial has lowest average loss")
print(f"{1+np.argmin(arr_rmse)} Degree polynomial has lowest rmse loss")
```
